```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/customer_shopping_behavior.csv')
```

```python
df.isnull().sum()
```

```python
#filling null values with category wise median (mean is effected by outliers)
df.groupby('Category').median(numeric_only=True)
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
#snack_casing
df.columns = (df.columns.str.strip().str.lower().str.replace(' ', '_'))
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.info()
```

```python
#create new column age_group
labels=['Young Adult','Adult','Middle-aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels) #sorts age values, divides then so each group has same number of rows
df[['age','age_group']].head(10)
```

```python
#create purchase_frequence_days
#creates a dictionary to map values
frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Annually':365,
    'Every 3 Months':90
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping).fillna(0).astype(int)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```
